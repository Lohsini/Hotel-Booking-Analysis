```python

def setAge():
  # 设置正态分布的均值和标准差
  mean_age = 35
  std_dev_age = 10

  # 生成随机年龄数据
  ages = np.random.normal(mean_age, std_dev_age, 1000)
  
  ages = np.rint(ages)
  # 确保所有年龄至少为18岁
  ages = np.maximum(ages, 18)
  

  print(ages)
setAge()
```
